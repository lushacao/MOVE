# Demo Quickstart
This notebook contains easily editable code to use MOVE. The example is currently loaded with example data but can be changed to user data.

### Optional parameters
User defined parameters. Change nothing to use the defaults. For more information on these parameters, visit the documentation notebook of MOVE.
```
species
threshold
figsize
```
### Mandatory parameter
```
case
control
df_common
df_unique
```

## How to create MOVE object
First we create the MOVE object.

````python
import MOVE
````

````python
mv = MOVE.move()
````

````python
#?mv
````

Then we read in the example data and add them to the MOVE

````python
import pandas as pd
````

````python
protein_df = pd.read_csv("tutorial_data/combined_protein.tsv",sep = '\t')
````

````python
sample_df = pd.read_excel("tutorial_data/user_sample_names.xlsx")
````

````python
peptide_df = pd.read_csv("tutorial_data/combined_peptide.tsv",sep = '\t')
````

````python
#?mv.add_dataframes
````

````python
mv.add_dataframes(protein_df, sample_df, peptide_df)
````

## How to normalize the protein data

````python
#?mv.log2_transform
````

````python
mv.log2_transform(figsize=(8,7))
````

````python
mv.median_normalize(figsize=(8,7))
````

## How to remove noise and select common proteins

````python
?mv.noise_filter
````

````python
df_common = mv.noise_filter()
````

````python
mv.plot_common_proteins(df_common,figsize = (6,8))
````

## How to generate Venn diagram for case vs control

````python
#?mv.unique_proteins
````

````python
H1975_unique = mv.unique_proteins(df_common, control = 'H1975_unst', case = 'H1975_TGFb')
````

````python
H838_unique = mv.unique_proteins(df_common, control = 'H838_unst', case = 'H838_TGFb')
````

## How to verify the peptides of the uniquely expressed proteins
Generate the dashboard

````python
#?mv.peptide_verification
````

````python
mv.peptide_verification(case = 'H1975_TGFb', df_unique = H1975_unique)
````

````python
mv.peptide_verification(case = 'H838_TGFb', df_unique = H838_unique)
````
